```python
plt.hist(df['Maths'], bins=7,color='blue', edgecolor='black')
plt.title('Distribution of Maths Scores')
plt.xlabel('Maths Scores')
plt.ylabel('Frequency')
plt.show()
#total range = max score - min score
#bins means wanting the total range in the specified interval in such way it gets divded
#here 14 is the total range
#when bins is 7 ...so there will be 7 intervals
#if bins is 1...then there will be a one range from 85 to 99
#total range/given bins = range btwn each interval
```

```python
df.groupby('Section')['Maths'].mean().plot(kind='bar', color='green')
plt.title('Average Maths Scores by Section')
plt.xlabel('Section')
plt.ylabel('Average Maths Score')
plt.show()
#bar chart for the new dataframe which is grouped based on section (average on maths[one attribute] for each section only)
```

```python
df.groupby('Section')['Maths'].mean()
df.plot(kind='bar', color='green')
plt.title('Students Data')
plt.xlabel('ID')
plt.ylabel('Range')
plt.show()
#bar plot based on entire dataframe (all attributes)
```

```python
plt.scatter(df['Maths'], df['Science'], color='purple')
plt.title('Maths vs. Science Scores')
plt.xlabel('Maths Scores')
plt.ylabel('Science Scores')
plt.show()
#normal like plotting points in x and y axis
```

```python
df['Maths'].value_counts().plot(kind='pie', autopct='%.2f%%')
plt.title('Maths Marks')
plt.show()
#counts no.of occurence for each value in Maths attribute according to that percentage is given
```

```python
df['Section'].value_counts().plot(kind='pie', autopct='%.2f%%') #autopct is in which format u want to display
plt.title('Proportion of Students in Each Section')
plt.show()
#no of occurence of each section which means gives no. of students in each section according to that percentage is given
```

```python
df['Name'].value_counts().plot(kind='pie', autopct='%.2f%%')
plt.title('Students')
plt.show()
#all names are occured only once...so equal percentage
```

```python
sns.boxplot(x='Section', y='Maths', data=df)
plt.title('Box Plot of Maths Scores by Section')
plt.xlabel('Section')
plt.ylabel('Maths Scores')
plt.show()
#its based on IQR
```

```python
sns.heatmap(df[['Maths', 'Science', 'Average']].corr(),annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
#corr means correlation gives matrix
#corr is found using certain formula
#annot to display the correlation coefficients inside the cell
#1 indicates a perfect positive correlation,0 indicates no correlation and -1 indicates a perfect negative correlation.
#cool colors represent negative correlations, and warm colors represent positive correlations.
```
